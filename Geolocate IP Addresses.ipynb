# IP Address Geolocation

```python
import pandas as pd
import requests

# Sample DataFrame with customer IPs
data = {'IP': ['8.8.8.8', '1.1.1.1']}
df = pd.DataFrame(data)

# Function to get geolocation
def get_geolocation(ip):
    access_token = 'your_ipinfo_access_token'
    url = f'https://ipinfo.io/{ip}/geo?token={access_token}'
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return data.get('loc', '0,0').split(',')
    else:
        return ['0', '0']

# Apply the function to the DataFrame
df[['Latitude', 'Longitude']] = df['IP'].apply(get_geolocation).apply(pd.Series)

print(df)
```

```python
import folium

# Create a map centered around the average latitude and longitude
map_center = [df['Latitude'].astype(float).mean(), df['Longitude'].astype(float).mean()]
customer_map = folium.Map(location=map_center, zoom_start=2)

# Add points to the map
for i, row in df.iterrows():
    folium.Marker(
        location=[float(row['Latitude']), float(row['Longitude'])],
        popup=row['IP']
    ).add_to(customer_map)

# Save map to an HTML file
customer_map.save('customer_map.html')
```

```python
import folium
import pandas as pd
from IPython.display import display
import requests

# Example dataframe with IP addresses and their corresponding geographical coordinates
df = pd.DataFrame({
    'ip': ['159.146.13.214', '2a02:ff0:220:6b2e:8c:46b2:7cf5:b3a4'],
})

# Your IPstack API key
api_key = '9225c622f5a01c42fda4c681b5577c39'

# Function to get latitude and longitude from IP
def get_coordinates(ip):
    response = requests.get(f'http://api.ipstack.com/{ip}?access_key={api_key}')
    data = response.json()
    if 'latitude' in data and 'longitude' in data:
        return data['latitude'], data['longitude']
    else:
        print(f"Error: {data}")
        return None, None

# Get coordinates for your IP addresses
coordinates = [get_coordinates(ip) for ip in df['ip']]

# Remove any None values from the list
coordinates = [(lat, lon) for lat, lon in coordinates if lat is not None and lon is not None]

# Update the dataframe with actual coordinates
if coordinates:
    df['lat'], df['lon'] = zip(*coordinates)

# Create a map centered around the average coordinates
map = folium.Map(location=[df['lat'].mean(), df['lon'].mean()], zoom_start=5)

# Add points to the map
for i, row in df.iterrows():
    folium.Marker(location=[row['lat'], row['lon']], popup=row['ip']).add_to(map)

# Display the map
display(map)
```
